```python
import torch
import torch.nn as nn
from transformers import AutoTokenizer, AutoConfig, AutoModelForCausalLM
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
import os

from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file

huggingface_api_key  = os.environ['HUGGINGFACE_KEY']
print(huggingface_api_key)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    "meta-llama/Llama-2-7b-chat-hf",
    device_map='cuda',
    cache_dir='./base_models',
    token=huggingface_api_key
)

tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf",
                                         cache_dir='./base_models.',
                                         token=huggingface_api_key)
```

```python
def get_llama2_chat_response(prompt, max_new_tokens=100):
  inputs = tokenizer(prompt, return_tensors="pt").to(device)
  outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, temperature=0.0001)
  response = tokenizer.decode(outputs[0], skip_special_tokens=True)
  return response
```

```python
prompt = '''
[INST]
Classify the text into neutral, negative or positive.
Text: ผมไม่ชอบสินค้านี้
[/INST]
'''

print(get_llama2_chat_response(prompt))
```

```python
prompt = '''
[INST]
what is the Thai translation of the text
Text: I like eating
[/INST]
'''
print(get_llama2_chat_response(prompt))
```
